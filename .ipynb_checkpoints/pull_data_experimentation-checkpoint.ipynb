Notice that the followerCount column has dictionaries inside it. If we want to separate out these dictionaries into their own column, we can use the `json_normalize` method

```python
pd.json_normalize(newdata['elements'][0]['followerCountsByStaffCountRange'])
```

```python
#Change the order of the columns of one of your tables
Countbyindustrycolumn.reindex(["Organic Follower Count", "Industry", "Paid Follower Count"], axis =1)
```

```python
#Display only the rows which fulfill a certain condition
Countbyindustrycolumn.loc[Countbyindustrycolumn["Organic Follower Count"] >= 50]
```

```python
#Add a new column to an existing table containing only zeroes
Countbyindustrycolumn["new"] = 0
Countbyindustrycolumn
```
